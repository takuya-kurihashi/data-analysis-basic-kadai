```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

# 折れ線
g1 = [1, 2, -5, 2]
axes[0,0].plot(g1)

# sin関数
gx2 = np.linspace(0, 10, 100)
gy2 = 2 + 2 * np.sin(2 * gx2)
axes[0,1].plot(gx2, gy2, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
         ylim=(-1, 5), yticks=np.arange(-1, 6))

# ヒストグラム
g3 = np.random.normal(15, 5, 2000)
axes[0,2].hist(g3)

# 散布図
np.random.seed(3)
gx4 = 4 + np.random.normal(0, 2, 60)
gy4 = 4 + np.random.normal(0, 2, len(gx4))
axes[1,0].scatter(gx4, gy4)

# ダミー
axes[1,1].plot()
axes[1,2].plot()
```
